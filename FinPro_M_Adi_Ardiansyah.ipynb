## Pendahuluan :
1. Latar Belakang
  - Kanker payudara adalah kanker yang terbentuk di jaringan payudara. Kanker payudara terjadi ketika sel-sel pada jaringan di payudara tumbuh secara tidak terkendali dan mengambil alih jaringan payudara yang sehat dan sekitarnya.
  - Pada 2022, terdapat sekitar 2,4 juta kasus baru kanker payudara, atau 11,6 persen dari total kasus kanker baru. Kendati demikian, kanker payudara merupakan jenis kanker yang paling banyak diidap oleh wanita. Kanker ini juga menyebabkan sekitar 670 ribu kematian, atau 6,9 persen dari total kasus kematian akibat kanker.
  - Meningkatkan pemahaman tentang faktor-faktor yang mempengaruhi kanker payudara

2. Tujuan Penelitian
  - Menganalisis hunbungan antara karakteristik pasien dan hasil diagnosis kanker payudara untuk mendapatkan prediksi apakah pasien terkena kanker atau tidak. Sehingga bisa segera mendapatkan penanganan lebih cepat dan mengurangi persentase kematian yang disebabkan olek kanker payudara.
3. Dataset <br>
  a. S/N = Unique identification for each patient.<br>
  b. Year=The year diagnosis was conducted. <br>
  c. Age = Age of patient at the time of diagnose. <br>
  d. Menopause = Whether the patient is pro or postmenopausal at the time diagnose,0 MEANS THAT THE PATIENT HAS REACHED MENOPAUSE WHILE 1 MEANS THAT THE PATIENT HAS NOT REACHED MENOPAUSE YET. 0 = Menopause , 1 = Non Menopause <br>
  e. Tumor size = The size in centimeter of the excised tumor. <br>
  f. Involved nodes = The number of axillary lymph nodes that contain metastatic,"CODED AS A BINARY DISTRI UTION OF EITHER PRESENT OR ASENT. 1 MEANS PRESENT, 0 MEANS ABSENT." mengecek kelenjar getah bening aksila yang mengandung Metastasis. 1 = ada, 0 = tidak ada<br>
  g. Breast = If it occurs on the left or right side,"CODED AS A BINARY DISTRIBUTION 1 MEANS THE CANCER HAS SPREAD, 0 MEANS IT HASN'T SPREAD YET."<br>
  h. Metastatic = If the cancer has spread to other part of the body or organ.1 = metastasis , 0 = non metastasis . Note : metastasis adalah penyebaran sel kanker dari lokasi asalnya(Tumor Primer)<br>
  i. Breast quadrant = The gland is divided into 4 sections with nipple as a central point.<br>
  j. History = If the patient has any history or family history on cancer,"1 means there is a history of cancer , 0 means no history."<br>
  k. Diagnosis result = Instances of the breast cancer dataset. Benign = Jinak atau sehat. Malignant = Ganas atau sakit

```python
cols_to_plot = ['Menopause', 'Inv-Nodes', 'Breast', 'Metastasis',
                'History', 'BreastQuadrant', 'DiagnosisResult']

def autopct_fun(abs_values):
    gen = iter(abs_values)
    return lambda pct: f"{pct:.1f}%\n({next(gen)})"

fig,ax = plt.subplots(nrows = 4, ncols = 2, figsize = (9,14))
ax = ax.flat

for i,col in enumerate(cols_to_plot):
    df_class = df_eda[col].value_counts().to_frame()
    labels = df_class.index
    values = df_class.iloc[:,0].to_list()
    ax[i].pie(x = values, labels = labels, autopct=autopct_fun(values), shadow = True, textprops = {'color':'white', 'fontsize':10, 'fontweight':'bold'})
    ax[i].legend(labels)
    ax[i].set_title(col, fontsize = 14, fontweight = "bold", color = "black")
    ax[i].axis('equal')

fig.delaxes(ax = ax[7])
fig.tight_layout()
fig.show()
```

### Memisahkan kolom DiagnosisResult dan hanya mengambil kolom Malignant(kanker)

```python
dr = ['DiagnosisResult']
df1 = pd.get_dummies(df1, columns = dr)
df1.info()
```

```python
#Kita akan drop kolom 'DiagnosisResult_Benign' karena kolom itu tidak kita gunakan sebagai feature
df1 = df1.drop('DiagnosisResult_Benign', axis = 1)

#Mengganti kolom 'DiagnosisResult_Malignant' menjadi 'Malignant'
df1.rename(columns={'DiagnosisResult_Malignant': 'Malignant'}, inplace=True)
df1.sample(10)
```

```python
from sklearn.model_selection import train_test_split

X = df1.drop(columns='Malignant')
y = df1[['Malignant']]
# y = y.values.reshape(len(y),)

X_train, X_test,y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

### Memeriksa Multicollinearity

```python
from statsmodels.tools.tools import add_constant
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

X = add_constant(X_train)
vif_ch = pd.DataFrame([vif(X.values, i)
                      for i in range (X.shape[1])],
                     index = X.columns).reset_index()
vif_ch = pd.DataFrame([vif(X.values, i) for i in range(X.shape[1])],
                      index=X.columns).reset_index()
vif_ch.columns = ['X','vif_score']
vif_ch = vif_ch.loc[vif_ch.X!='const']
vif_ch
```

## KNN Model

```python
# define the estimator/model
from sklearn.neighbors import KNeighborsClassifier

knn_clf = KNeighborsClassifier()
```

```python
# hyperparameter tuning
from sklearn.model_selection import GridSearchCV

parameters = {
    'n_neighbors': (2,3,4,5,6,7,8)
}

knn_clf_gridcv = GridSearchCV(knn_clf, parameters, cv=5, scoring='recall')
knn_clf_gridcv.fit(X_train, y_train)

# the compact results
cv_result = pd.DataFrame(knn_clf_gridcv.cv_results_)
retain_cols = ['params','mean_test_score','rank_test_score']
cv_result[retain_cols]
```

```python
# what if we also want to tune "weights" hyperparameter
# "uniform": all K neighbors have the same influence
# "distance": closer neighbor in those K neighbors has stronger influence
from sklearn.model_selection import GridSearchCV

parameters = {
    'n_neighbors': (2,3,4,5,6,7,8),
    'weights':('uniform','distance')
}

# now we use recall (sensitivity) as metric
knn_clf_gridcv = GridSearchCV(knn_clf, parameters, cv=5, scoring='recall')
knn_clf_gridcv.fit(X_train, y_train)

# the results
cv_result = pd.DataFrame(knn_clf_gridcv.cv_results_)
retain_cols = ['params','mean_test_score','rank_test_score']
cv_result[retain_cols].sort_values('rank_test_score')

# classification report
from sklearn.metrics import classification_report
y_pred = knn_clf_gridcv.best_estimator_.predict(X_test)
print(classification_report(y_test, y_pred))
```

Hasil Model KNN : <br>
Recall = 0.84<br>
f1-score = 0.84<br>
time_processing = 2s
